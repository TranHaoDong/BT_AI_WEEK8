```python
from keras.models import load_model
model = load_model('/content/drive/MyDrive/Colab Notebooks/W10_Cifar100.h5')
```

```python
import matplotlib.pyplot as plt
from keras.utils import load_img, img_to_array
img= load_img('/content/drive/MyDrive/Colab Notebooks/Hinh_Anh/A20.jpg', target_size= (32,32))
plt.imshow(img)
plt.show()
```

```python
import numpy as np
img = img_to_array(img)
img= img.reshape(1, 32*32*3)
img= img.astype('float32')
img= img/255
np.argmax(model.predict(img), axis= -1)
```
